```python
from scipy import stats
from pycountry_convert import country_alpha2_to_continent_code, country_name_to_country_alpha2
```

```python
#Bubble chart to  visualize Life Expectancy vs GDP Plot. 
#Where size of the bubble determines the population of that particular country and 
#color denotes the continent of the country.
continents = {
    'NA': 'North America',
    'SA': 'South America', 
    'AS': 'Asia',
    'OC': 'Australia',
    'AF': 'Africa',
    'EU': 'Europe'
}
continent = []
for country in life_span['Country']:
    try:
        continent.append(continents[(country_alpha2_to_continent_code(country_name_to_country_alpha2(country)))])
    except:
        continent.append("Africa")
    
life_span["Continent"] = continent
to_bubble = life_span[["Country", "Year", "Life expectancy ", "GDP", "Population", "Continent"]]
```

```python
to_bubble
```

```python
#Countries with high GDP(gross domestic product) also have a high life expectancy rate than countries with low GDP.
#most African countries have low expectancy rate.
#Continent codes were created and the countries in the table were then grouped by continent.
```

```python
d = to_bubble.drop('Year', axis=1)
to_bubble = d.groupby(['Country', 'Continent']).mean().reset_index()
to_bubble['GDP'] = np.log10(to_bubble['GDP'])
to_bubble.head()
```

```python
#Identified outliers with the Box plot, not it is time to remove them.
Q1 = life_span[cols].quantile(0.25)
Q3 = life_span[cols].quantile(0.75)
IQR = Q3 - Q1

outliers = ((life_span[cols] < (Q1 - 1.7 * IQR)) | (life_span[cols] > (Q3 + 1.7 * IQR))).any(axis=1)

life_span = life_span[~outliers]

life_span.head()
```

```python
#Infant/adult mortality, and disease all have negative correlation with life expectancy while
#there is a positive correlation for those that have been vaccinated, have higher income, and schooling.
l= [life_span['Life expectancy '].dropna(), life_span['Income composition of resources'].dropna(), life_span['Schooling'].dropna()]
for i in l:
    plt.figure(figsize=(20,5))
    sns.histplot(i, kde=True, color="orange")
```

```python
nr_rows = 5
nr_cols = 4

fig, axs = plt.subplots(nr_rows, nr_cols, figsize=(nr_cols*5,nr_rows*5))
numerical_feats = life_span.dtypes[life_span.dtypes != "object"].index
li_num_feats = list(numerical_feats)
li_not_plot = []
li_plot_num_feats = [c for c in list(numerical_feats) if c not in li_not_plot]
life_span.dropna(inplace = True)

for r in range(0,nr_rows):
    for c in range(0,nr_cols):  
        i = r*nr_cols+c
        if i < len(li_plot_num_feats):
            sns.regplot(x = life_span[li_plot_num_feats[i]],y = life_span["Life expectancy "] , color = 'blue', ax = axs[r][c])
            stp = stats.pearsonr(life_span[li_plot_num_feats[i]], life_span["Life expectancy "])
            str_title = "r = " + "{0:.2f}".format(stp[0]) + "      " "p = " + "{0:.2f}".format(stp[1])
            axs[r][c].set_title(str_title,fontsize=11)
            
            
plt.tight_layout()    
sns.set(color_codes=True)
```

```python
#I removed all the countries with 0 alcohol consumption. 

dry_country = alcohol_consumption.columns[1:-1]  # assuming 'Country' is the first column and 'Continent' is the last

# Create a boolean mask where each cell is True if the value is 0
mask = alcohol_consumption[dry_country].eq(0).all(axis=1)

# Use the mask to filter out rows with all zeros
filtered_alc_con = alcohol_consumption[~mask]
```

```python
filtered_alc_con #using the filtered DF moving forward.
```

```python
#My next goal is to add the consumption amount for all the alcohol types into one column and order
#by the country in DESC order.

# Create a new column 'Total' which is the sum of the specified columns
filtered_alc_con['Total'] = filtered_alc_con[['beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol']].sum(axis=1)
```

```python
sorted_df = filtered_alc_con.sort_values(by='Total', ascending=False)
```
